```python
def cleanYear(data):
    i=1950
    for x in range(151):
        data['Year'][x]=i
        i=i+1
    return data
def load_and_process(path):
    past=(
        pd.read_csv(path)
        .iloc[:55]
        .dropna(axis='columns', how='any')
        .rename(columns={"DateTime": "Year"})
    )
    future1=(
        pd.read_csv(path)
        .iloc[56:151]
        .dropna(axis='columns', how='any')
        .rename(columns={"DateTime": "Year"})
    )
    future2=(
        future1
        .assign(Comparison=future1['RCP 8.5 Median']-future1['RCP 2.6 Median'])
    )
    past=cleanYear(past)
    future=cleanYear(future2)
    return (past,future2)
```

```python
#select only the median columns to have a better look
years=futureTemp['Year']
plt.plot(years,futureTemp['RCP 2.6 Median'],label='RCP 2.6')
plt.plot(years,futureTemp["RCP 4.5 Median"],label='RCP 4.5')
plt.plot(years,futureTemp['RCP 8.5 Median'],label='RCP 8.5')
plt.legend()
plt.ylabel('Degree')
plt.xlabel('Year')
```

There is something interesting while working with the graph of future prediction. I found that the growth rate is not the same for all RCPs. The larger the RCP the faster the temperature will grow.

```python
years=futureTemp['Year']
plt.plot(years,futureTemp['Comparison'],label='Comparison of RCP 8.5 and 2.6')
plt.legend()
plt.ylabel('Degree')
plt.xlabel('Year')
```

The graph shows the difference of the temperature change between RCP 8.5 and RCP 2.6. We can see that with the predictions go further, the larger the difference between the mean temperature of RCP 8.5 and RCP 2.6 would be.
