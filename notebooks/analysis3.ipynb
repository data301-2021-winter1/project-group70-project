```python
import pandas as pd
hottestday = pd.read_csv ('../data/raw/Hottest_Day.csv')
i=1950
for x in range(151):
    hottestday["DateTime"][x]=i
    i=i+1

hotday_proj = hottestday[56:151]
hotday_hist = hottestday[0:55]

hotday_proj_clean=hotday_proj.dropna(axis='columns',how='all')
hotday_proj_clean.pop('RCP 2.6 Range (low)')
hotday_proj_clean.pop('RCP 2.6 Range (high)')
hotday_proj_clean.pop('RCP 4.5 Range (low)')
hotday_proj_clean.pop('RCP 4.5 Range (high)')
hotday_proj_clean.pop('RCP 8.5 Range (low)')
hotday_proj_clean.pop('RCP 8.5 Range (high)')
hotday_hist_clean=hotday_hist.dropna(axis='columns',how='all')
hotday_hist_clean.pop('Historical Range (low)')
hotday_hist_clean.pop('Historical Range (high)')
hotday_hist_clean
```

```python
#step 3
def load_and_process(path='../data/raw/Hottest_Day.csv' ):
    # Method Chain 1 (hist data processing)

    hist= (
         pd.read_csv(path)
      
           [0:55]
            .dropna(axis='columns',how='all')
            .drop(columns=["Historical Range (high)","Historical Range (low)"])
      
      )

    # Method Chain 2 (projected data processing)

    proj= (
         
          pd.read_csv(path='../data/raw/Hottest_Day.csv')
      
           [56:151]
            .dropna(axis='columns',how='all')
            .drop(columns=["RCP 8.5 Range (high)","RCP 8.5 Range (low)","RCP 4.5 Range (high)","RCP 4.5 Range (low)","RCP 2.6 Range (high)","RCP 2.6 Range (low)"])
        
      )


    return (hist, proj) 
```
